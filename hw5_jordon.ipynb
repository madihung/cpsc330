```python
#1. 
numeric_features=["LIMIT_BAL","AGE","BILL_AMT1","BILL_AMT2","BILL_AMT3","BILL_AMT4","BILL_AMT5","BILL_AMT6","PAY_AMT1","PAY_AMT2","PAY_AMT3","PAY_AMT4","PAY_AMT5","PAY_AMT6"]
categorical_features=["MARRIAGE","PAY_0","PAY_1","PAY_2","PAY_3","PAY_4","PAY_5","PAY_6"]
ordinal_features=["EDUCATION"]
drop_features=["ID","SEX"]
target_column="default.payment.next.month"

education_levels = ["graduate school", "university", "high school", "others", "unknown", "unknown"]


# make transformers
numeric_transformer = make_pipeline(StandardScaler())

ordinal_transformer = make_pipeline(
    OrdinalEncoder(categories=[education_levels], dtype=int)
)

categorical_transformer = make_pipeline(
    OneHotEncoder(handle_unknown="ignore", sparse=False),
)

#2. make preprocessor/column transformer
preprocessor = make_column_transformer(
    (numeric_transformer, numeric_features),
    (ordinal_transformer, ordinal_features),
    (categorical_transformer, categorical_features),
    ("drop", drop_features),
)

#create feature/target data
X_train,y_train=train_df.drop(columns=[target_column]),train_df[target_column]
X_test,y_test=test_df.drop(columns=[target_column]),test_df[target_column]

#3. make preprocessor pipeline 
lr_pipe=make_pipeline(preprocessor,LogisticRegression())
```
